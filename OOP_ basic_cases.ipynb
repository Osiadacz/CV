```python
class Person(object):
    def __init__(self,name,lastname):   
        self.name = name
        self.lastname = lastname
        
        
    def __str__(self):
        return "ssn: {}, name: {}, lastname: {}, address: {}".format(self.ssn, self.name, self.lastname, self.address)
       
        
class Parent(Person):
    def __init__(self,name,lastname,ssn,address):   
    
        Person.__init__(self, name, lastname)  
        self.ssn = ssn
        self.address = address

parent_list = []
parent_dict = {}


Parent_1 = Parent(input("Enter the name of the parent: "),     
                     input("Enter the lastname of the parent: "),
                     input("Enter the ssn of the parent: "),
                     input("Enter the address of the parent: "),)


parent_dict[Parent_1.ssn] = Parent_1



Parent_2 = Parent(input("Enter the name of the parent: "),     
                     input("Enter the lastname of the parent: "),
                     input("Enter the ssn of the parent: "),
                     input("Enter the address of the parent: "),)


parent_dict[Parent_2.ssn] = Parent_2



parent_list.append(parent_dict.copy())

information = lambda: print(Parent_1)   
information()
information = lambda: print(Parent_2)   
information()


for i in parent_list:   
    print(i)

class Student(Parent):
    def __init__(self,name,lastname,ssn,address,year,subjects,extra):  
    
        Parent.__init__(self, name, lastname,ssn,address)    
        self.year = year
        self.subjects = subjects
        self.extra = extra
        
    def __str__(self): 
        return "ssn: {}, name: {}, lastname: {}, address: {}, year: {}, subjects: {}, extracurricular: {}".format(self.ssn,
                                                                                                                                        self.name,
                                                                                                                                        self.lastname, self.address,
                                                                                                                                        self.year, self.subjects, self.extra)

student_list = []
student_dict = {}


Student_1 = Student(input("Enter the name of the Student: "),       
                     input("Enter the lastname of the Student: "),
                     input("Enter the ssn of the Student: "),
                     input("Enter the address of the Student: "),
                     input("Enter the year attended by the Student: "),
                     input("Enter the subjects taken by the Student: "),
                     input("Enter the extracurricular taken, if apply: "),)


student_dict[Student_1.ssn] = Student_1

 
Student_2 = Student(input("Enter the name of the Student: "),   
                     input("Enter the lastname of the Student: "),
                     input("Enter the ssn of the Student: "),
                     input("Enter the address of the Student: "),
                     input("Enter the year attended by the Student: "),
                     input("Enter the subjects taken by the Student: "),
                     input("Enter the extracurricular taken, if apply: "),)


student_dict[Student_2.ssn] = Student_2


student_list.append(student_dict.copy())

information = lambda: print(Student_1)   
information()
information = lambda: print(Student_2)   
information()

for i in student_list:    
    print(i)
```
